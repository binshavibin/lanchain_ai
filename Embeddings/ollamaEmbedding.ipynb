####Embedding by using models from ollama
Ollama Embeddings



```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embedding = (
    OllamaEmbeddings(model="gemma:2b")
    )
embedding
```

```python
r1 = embedding.embed_documents([
    "Hello world",
    "Ollama is a great tool for embeddings"
])
r1
```

```python
query = embedding.embed_query("What is Ollama?")
query
```

```python
embedding = (
    OllamaEmbeddings(model="mxbai-embed-large")
    )
text = "What is the capital of France?"
embedding.embed_query(text)
```
