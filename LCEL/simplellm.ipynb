```python
import os
import openai
from dotenv import load_dotenv

load_dotenv()
groq_api_key = os.getenv("GROQ_API_KEY")
openai.api_key = os.getenv("OPEN_AI_API_KEY")
```

```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage

# Initialize model
model = ChatGroq(model="llama-3.1-8b-instant", groq_api_key=groq_api_key)

# Messages
message = [
    SystemMessage(content="Translate the following from English to hindi."),
    HumanMessage(content="What is the capital of France?")
]

response = model.invoke(message)
response
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
parser.invoke(response)
```

```python
##Using LCEL we can chain of component
chain = model | parser
chain.invoke(message)
```

```python
##Using Promt template
from langchain_core.prompts import ChatPromptTemplate   

generic_prompt = "Translate the prompt to {language}"

prompt = ChatPromptTemplate.from_messages(
    [
      ( "system", generic_prompt),
        ( "user", "{text}")
    ]
)
response = prompt.invoke({"language": "Hindi", "text": "Hello"})
response.to_messages()
```

```python
##Chaning together with LCEL
chain  = prompt | model | parser
chain.invoke({"language": "Hindi", "text": "Hello"})
chain.invoke({"language": "malayalam", "text": "Good Morning"})
```
